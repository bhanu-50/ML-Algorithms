```
w = np.random.normal(size=(2,))
```

```
w0 = np.random.normal()
```

```
w0
```

```
def display(w, w0, X, Y):
  intercept= -w0 /w[1]
  slope = -w[0]/w[1]
  hplane_x = np.arange(0,15)
  hplane_y = slope * hplane_x + intercept
  plt.plot(X[:, 0][Y==-1], X[:, 1][Y==-1],'g^')
  plt.plot(X[:, 0][Y==1],X[:, 1][Y==1],'bs')
  plt.plot(hplane_x,hplane_y,'r-')
  plt.xlabel("X1")
  plt.ylabel("X2")
  plt.show()
```

```
display(w,w0,X,Y)
```

```
def persptron(X,Y,w,w0,lr = 0.5):
    n = len(X)
    for iteration in range(1001):
      misclassifed = 0
      for i in range(n):
        half_space = np.dot(w,X[i])+w0
        if  half_space*Y[i]<0:
          misclassifed += 1
          w = w + lr * Y[i] * X[i]
          w0 = w0 + lr * Y[i]
      acc = (n-misclassifed)/n
      if iteration %500 == 0:
        display(w, w0, X, Y)
        print(f"accuracy{acc}")
persptron(X,Y,w,w0)
```
