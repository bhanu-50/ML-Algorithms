```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import math
import plotly.express as px
warnings.filterwarnings('ignore')
```

```
X = np.linspace(-10,10,200)
y = 2.4*X + 4 + np.random.normal(0,5,len(X))
```

```
px.scatter(X,y)
```

```
def plot_line(w,w0,X,y):
  plt.scatter(X,y)
  plt.plot(X,w*X+w0,color='red')
  plt.show()
```

```
plot_line(2.4,4,X,y)
```

```
#Loss_Function
def loss_function(w,w0,X,y):
  predict = w*X + w0
  return np.sum((y-predict)**2)
```

```
def gradient(w,w0,X,y):
  n = len(X)
  predict = w*X + w0
  return -np.sum(2*X*(y-predict))/n, -np.sum(2*(y-predict))/n
```

```
def gradient_decent(w,w0,X,y,learning_rate=0.01,ITR = 1000):
  for i in range(ITR):
    gradw, gradw0 = gradient(w,w0,X,y)
    w -= learning_rate * gradw
    w0 -= learning_rate * gradw0
    print(w,w0)
    if abs(gradw) < 0.01 and abs(gradw0) <0.01:
      plot_line(w,w0,X,y)
      print(f'{w},{w0}')
      break
```

```
gradient_decent(100,200,X,y)
```
