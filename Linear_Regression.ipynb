```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import plotly.express as px
warnings.filterwarnings('ignore')
```

```
X = np.linspace(-10,10,200)
y = 2.4*X + 4 + np.random.normal(0,2,len(X))
```

```
px.scatter(X,y)
```

```
def plot_line(w,w0,X,y):
  plt.scatter(X,y)
  plt.plot(X,w*X+w0,color='red')
  plt.show()
```

```
plot_line(2.4,4,X,y)
```
